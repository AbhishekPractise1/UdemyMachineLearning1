# Matplotlib example workflow


```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig , axes = plt.subplots(figsize = (5,5)) ##figsize = (width, height) of the figure
#Create a figure and a set of subplots

# 3. Plot data 
axes.plot(x,y) ##adding data to our axes

#4. Customize plot
axes.set(title = "Simple plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5.Save & show (you save the whole figure)
fig.savefig("DemoPlot1.png")

plt.show()
```

```python
import numpy as np
```

```python
#Creating some data
x = np.linspace(0,10,100)#Return evenly spaced numbers over a specified interval.
x[:10]
```

```python
#plotting the data and create a LINE PLOT
fig , ax = plt.subplots()
ax.plot(x,x**2);

##Default Matplotlib is a line plot
```

```python
#Use the same data and create a SCATTER PLOT
fig , ax = plt.subplots()
ax.scatter(x,np.exp(x));
#scatter plots = lot of dots
```

```python
#scatter plot exp 2
fig , ax = plt.subplots()
ax.scatter(x,np.cos(x));
```

```python
# PLot From Dictionary 
nutt = {"Almond Butter" : 10,
       "Peanut Butter" : 8,
       "cashew Butter" : 12}
fig , ax = plt.subplots()
ax.bar(nutt.keys(), nutt.values());
#ax.bar(x,y) here x cannot have positional argument but y can have 
ax.set(title = "Sales Graph",ylabel = "Prices($)");
#here ; removes the line of code that apperars above the plot
```

```python
fig ,ax = plt.subplots()
ax.barh(list(nutt.keys()) , list(nutt.values()));
#barh = horizontal bar
#It does not accepts dictionary as plotting values
#you have to convert it to lists
```

```python
#Make some data for Histogram and plot it
x = np.random.randn(1000)#Return a sample (or samples) from the "standard normal" distribution.
#Normal Distribution  = Kind of like symmetrical curve
fig,ax = plt.subplots()
ax.hist(x);
```

# Two optional for subplots(More than one plot in a cell)

```python
#Subplot option 1
fig,((ax1,ax2),(ax3,ax4))  = plt.subplots(nrows = 2,
                                         ncols = 2,
                                         figsize = (10,5))
#nrows = number of rows of subplot grid
#ncols = number of column of subplot grid
#figsize = (width , height of each figure/plot)

#PLot to each different axis
ax1.plot(x,x/2);#top -left
ax2.scatter(np.random.random(10),np.random.random(10));#top-left
ax3.bar(nutt.keys(),nutt.values());#bottom-right
ax4 . hist(np.random.randn(1000));#bottom-left
```

```python
#Subplot Option 2 :
fig , ax = plt.subplots(nrows = 2 , ncols = 2 , figsize = (10,5))

#plot to each different index
ax[0,0] .plot(x,x/2);#top -left
ax[0,1] . scatter(np.random.random(10),np.random.random(10));#top-left
ax[1,0] .bar(nutt.keys(),nutt.values());#bottom-right
ax[1,1] .  hist(np.random.randn(1000));#bottom-left
#ax[row , columns]
```

# Data visualization Tip
This is a common mistake with beginners.

That is we assume that a graph is always telling the truth.

And we jump to conclusions right away from data.

Instead we want to pause and reflect and make sure that whatever we're visualizing is accurate and testing

it with different methods trying different visualizations.

We don't want to jump to conclusions right away from data.

It takes a lot to understand this data.

So just a quick note that is just because we have data doesn't mean we have knowledge.
